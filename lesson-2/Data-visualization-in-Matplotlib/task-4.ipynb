```python
import pandas as pd
import matplotlib.pyplot as plt

%config InlineBackend.figure_format='svg'
```

```python
df = pd.read_csv('creditcard.csv')
print(df.head())
```

```python
df["Class"].value_counts().plot(kind='bar')
df["Class"].value_counts().plot(kind='bar', logy=True, title='graphic')
```

```python

plt.hist(df.loc[df['Class']==0, 'V1'], label='Class 0', log=True, density=True, bins=20, alpha=0.5, color='grey')
plt.hist(df.loc[df['Class']==1, 'V1'], label='Class 1', log=True, density=True, bins=20, alpha=0.5, color='red')
title_dict = {'fontsize': 14, 'color': 'lightblue', 'family':'calibri', 'fontweight':'bold'}
plt.title('graphic', fontdict=title_dict )
label_dict = {'family': 'calibri', 'size':10}
plt.legend(loc='upper center', frameon=False, prop=label_dict)
plt.xlabel('V1', color='lightblue')
plt.ylabel('density', color='lightblue')

```
